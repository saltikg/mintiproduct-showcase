# DuckDB Exploration (Safe Example)

This notebook demonstrates a small, safe example of how MintiProduct uses DuckDB for analytical queries.

No production data is included.

```python
import duckdb
con = duckdb.connect()
```

## Create Example Table

```python
con.execute("""
CREATE TABLE trends_demo AS
SELECT * FROM (
  VALUES
    ('winter skincare', 85),
    ('gift ideas', 92),
    ('cozy home decor', 77)
) AS t(topic, score);
""")
```

## Query the Table

```python
con.execute("SELECT * FROM trends_demo ORDER BY score DESC").df()
```
